```python
import matplotlib.pyplot as plt
```

```python
labels = ['Codar', 'Estudar', 'Exercitar', 'Ler']
vals = [7, 4, 5, 2]
colors = ["seagreen", "#db5e5e", "tab:blue", "tab:orange"]
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

explode = (0.1, 0, 0, 0)

ax.pie(
    vals, labels=labels, autopct="%.1f%%", shadow=True, explode=explode,
    startangle=60, colors=colors
)
ax.set_title('Tempo por Tarefa', fontsize=16)

plt.show()
```
