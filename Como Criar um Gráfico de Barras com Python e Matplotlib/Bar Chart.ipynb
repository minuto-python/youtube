```python
import matplotlib.pyplot as plt
```

```python
labels = ['Azul', 'Vermelho', 'Verde', 'Amarelo']
vals = [53, 83, 92, 65]
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(labels, vals, color=['blue', 'red', 'green', 'y'])

ax.set_title('Desafio', fontsize=18)

ax.set_ylabel('Pontuação', fontsize=14)
ax.set_xlabel('Times', fontsize=14)

for idx, val in enumerate(vals):
    txt = f'{val} pts'
    y_coord = val + 0.7
    x_coord = idx - 0.15
    
    ax.text(x=x_coord, y=y_coord, s=txt, fontsize=12)

ax.grid()

plt.show()
```
